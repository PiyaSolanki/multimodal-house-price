```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load data
train_df = pd.read_csv("../data/raw/train.csv")

print(train_df.shape)
train_df.head()
```

```python
train_df.info()
```

```python
train_df.isnull().sum().sort_values(ascending=False).head(10)
```

```python
plt.figure(figsize=(8, 5))
sns.histplot(train_df["price"], bins=50)
plt.title("House Price Distribution")
plt.show()
```

```python
import numpy as np

plt.figure(figsize=(8, 5))
sns.histplot(np.log1p(train_df["price"]), bins=50)
plt.title("Log-Transformed House Price Distribution")
plt.show()
```

```python
plt.figure(figsize=(12, 8))
corr = train_df.select_dtypes(include="number").corr()
sns.heatmap(corr, cmap="coolwarm", center=0)
plt.title("Feature Correlation Heatmap")
plt.show()
```

```python
plt.figure(figsize=(7, 6))
plt.scatter(train_df["long"], train_df["lat"], c=train_df["price"], s=5)
plt.colorbar(label="Price")
plt.title("House Prices by Location")
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.show()
```
